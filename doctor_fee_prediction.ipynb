```python
#Extracting relevant data from experience
def changeexperience(x):
    bans=x.split(' ')
    return bans[0]
```

```python
#making function to split the qualification data to split data by , or space
def changequal(x):
    bans=x.split(' ')
    sans=x.split(',')
    if ',' not in x:
        return bans[0]
    else:
        return sans[0]
```

```python
train["Qualification"]=train["Qualification"].apply(changequal)
```

```python
#checking the qualification data we can find out that there "-" present in few rows
train.Qualification.head(5)
```

```python
def remove_hyp(x):
    cabs=x.split('-')
    if '-'  in x:
        return cabs[0]
    else:
        return x
        
    
```

```python
train["Qualification"]=train["Qualification"].apply(remove_hyp)
```
